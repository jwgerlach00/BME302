```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_excel('Inulin_data.xlsx')

px.line(df, x='Time (min)', y='Inulin Concentration (cpm/mL)', title='Inulin vs Time', width=1000, height=800)
```

```python
# general flow rate

# GFR = excreation rate of inulin / conc of inulin in plasma
c_total_0 = 882
GFR = 2.76e3 / c_total_0  # cpm/min / cpm/mL = mL/min | When rat begins to excreate, bc not absorbed or excreated
GFR = GFR * 60 / 392  # mL/g-hours

k, _ = -np.polyfit(df['Time (min)'][6:], np.log(df['Inulin Concentration (cpm/mL)'][6:]), 1)

# c_total = 271  # polyfit time over 
# c_total_0

c_total = lambda t: c_total_0 * np.exp(-k * t)

time = np.linspace(0, 30, 1000)
px.line(x=time, y=[c_total(t) for t in time], width=1000, height=800)




# polyfit of naturallog and slope

# d take c and plug back in
```
