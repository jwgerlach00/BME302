```python
import pandas as pd
import plotly.express as px
```

```python
tidal_df = pd.read_excel('trial_data/use/volume_trial2.xlsx')
```

```python
crop_tidal_df = tidal_df.where(tidal_df.time > 50).dropna(subset=['time'])
```

```python
# Flow
f_fig = px.line(crop_tidal_df, x='time', y='flow', width=600, height=300)
f_fig.update_layout(
    title=dict(text='Normal Breathing Flow', x=0.5, font_size=40),
    xaxis=dict(title_text='Time (s)', title_font_size=30, tickfont_size=20),
    yaxis=dict(title_text='Flow (L/s)', title_font_size=30, tickfont_size=20),
)
f_fig.update_traces(
    line=dict(width=7)
)
f_fig.show()

# Volume
v_fig = px.line(crop_tidal_df, x='time', y='volume', width=600, height=300)
v_fig.update_layout(
    title=dict(text='Normal Breathing Volume', x=0.5, font_size=40),
    xaxis=dict(title_text='Time (s)', title_font_size=30, tickfont_size=20),
    yaxis=dict(title_text='Volume (L)', title_font_size=30, tickfont_size=20),
)
v_fig.update_traces(
    line=dict(width=7, color='red')
)
v_fig.show()
```

```python
norm_df = pd.read_excel('exercise3/data/volume_trial3.xlsx')
obstruct_df = pd.read_excel('exercise4/data/volume_trial2.xlsx')
norm_df, obstruct_df = [df.where(df.time > 25).dropna(subset=['time']) for df in (norm_df, obstruct_df)]


```

```python
import plotly.graph_objects as go

obstruct_fig = go.Figure(
    layout=go.Layout(width=800, height=500)
)
obstruct_fig.add_trace(go.Scatter(x=norm_df.time, y=norm_df.volume, name='Normal'))
obstruct_fig.add_trace(go.Scatter(x=obstruct_df.time, y=obstruct_df.volume, name='Obstructed'))
obstruct_fig.update_layout(
    title=dict(text='Obstructed vs. Normal Breathing', x=0.5, font_size=40),
    xaxis=dict(title_text='Time (s)', title_font_size=30, tickfont_size=20),
    yaxis=dict(title_text='Volume (L)', title_font_size=30, tickfont_size=20),
    legend=dict(title='Signal', font_size=20)
)
obstruct_fig.update_traces(
    line=dict(width=7)
)
obstruct_fig.show()
```
